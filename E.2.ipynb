#### En este ejercicio se debera importar numpy, scipy, pandas y matplotlib.pyplot

```python
import numpy as np
import scipy as stats
import pandas as pd
import matplotlib.pyplot as plt
```

 ##### Con pd.read_csv se llamaran los datos 'CasosGeneroEtario.csv' y lo guardaremos en la variable casosS

```python
casosS = pd.read_csv('CasosGeneroEtario.csv')
casosS
```

```python
casosS.dtypes
```

#### En este caso, la parte una nos pide graficar los datos acumulados por sexo de casos en el tiempo, con .drop eliminaremos 'Grupo de edad' ya que no nos piden esos datos, además con set_index colocaremos 'Sexo' como indice 

```python
casosS.drop(columns='Grupo de edad', inplace=True)
casosS.set_index('Sexo', inplace=True)
casosS
```

#### En el ejeX con el np.arange y len nos ayuda a grafica ya que con len que es un tipo de dato que devuelve la cantidad de artículos en un contenedor, asi que devolvera la columa 'Sexo'

```python
ejeX=np.arange(len(casosS.columns),step=1)
xlabels=list (casosS.columns)
```

```python
casosS.T.plot(figsize=(100,10), xlabel="FECHAS POR SEMANA", ylabel=" CANTIDAD DE CASOS" , 
         title = "Gráficos por grupo etario sin importar sexo en el tiempo", xticks=ejeX, rot=90)
```
